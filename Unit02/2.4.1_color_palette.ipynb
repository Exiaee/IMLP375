```python
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline 
import warnings
import pandas as pd
warnings.filterwarnings('ignore')
                        
current_palette = sns.color_palette() 
sns.palplot(current_palette)
sns.set(color_codes = True)
```

```python
#seaborn提供的6種調色板 
theme_list=['deep', 'muted', 'pastel','bright', 'dark', 'colorblind'] 
for i in theme_list: 
    sns.palplot(sns.color_palette(i))

```

```python
x, y = np.random.multivariate_normal([0, 0], [[1, -.5], [-.5, 1]], size=600).T
pal = sns.dark_palette("green", as_cmap=True)
pdx = pd.DataFrame(x)
pdy = pd.DataFrame(y)
pdz = pd.concat([pdx, pdy], axis=1)
pdz.columns=['X','Y']
#sns.kdeplot(x, y, cmap=pal)
sns.kdeplot(data=pdz, x='X', y='Y', cmap=pal)
```

```python
x1, y1 = np.random.multivariate_normal([0, 0], [[1, -.5], [-.5, 1]], size=600).T 
pal = sns.cubehelix_palette(light=1, as_cmap=True)
pdx1 = pd.DataFrame(x1)
pdy1 = pd.DataFrame(y1)
pdz1 = pd.concat([pdx1, pdy1], axis=1)
pdz1.columns = ['X','Y']
#sns.kdeplot(x, y, cmap=pal, shade=True)
sns.kdeplot(data=pdz1, x='X',y='Y',cmap=pal, fill=True)
```
